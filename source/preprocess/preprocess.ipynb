### 2.4. Xử lý dữ liệu không toàn vẹn
Dữ liệu có 4 dòng bị thiếu giá trị (276556, 285435, 323136, 262292), có thể xử lý bằng cách bỏ đi các dòng này do số dòng bị thiếu ít.

```python
null_positions = {
    col: dataFrame.index[dataFrame[col].isna()].tolist()
    for col in dataFrame.columns
    if dataFrame[col].isna().any()
}

for col, idx_list in null_positions.items():
    print(f"{col}: {idx_list}")
```

```python
print(dataFrame.loc[276556], "\n\n", dataFrame.loc[262292], "\n\n", dataFrame.loc[285435], "\n\n", dataFrame.loc[323136])
```

```python
dataFrame.notna().all().all()
```

```python
dataFrame = dataFrame.drop(index=[262292, 276556, 285435, 323136])
```

```python
dataFrame.notna().all().all()
```

### 2.5. Xử lý các cột không đúng kiểu dữ liệu
Có một số cột chứa nhiều giá trị số nguyên bị ép về kiểu số thập phân, cần biến đổi bằng cách ép lại về kiểu số nguyên.
Tuy nhiên, có một số cột (CWE Flag Count, Fwd Avg Bytes/Bulk) chỉ có phân phối 1 loại giá trị sau khi ép kiểu về,
có thể bỏ đi các cột này do dư thừa.

```python
object_cols = dataFrame.select_dtypes(include="object")
object_cols.head()
```

```python
dataFrame["CWE Flag Count"].value_counts().to_dict()
```

```python
dataFrame["Fwd Avg Bytes/Bulk"].value_counts().to_dict()
```

```python
dataFrame["Down/Up Ratio"].value_counts().to_dict()
```

```python
dataFrame = dataFrame.drop(columns=["CWE Flag Count", "Fwd Avg Bytes/Bulk"])
dataFrame["Down/Up Ratio"] = pd.to_numeric(dataFrame["Down/Up Ratio"], errors='coerce').astype(int) # ép về int
```

```python
dataFrame["Down/Up Ratio"].value_counts().to_dict()
```

### 2.6. Xóa cột Flow ID
Cột này chỉ có giá trị định danh nên cần được xóa khỏi tập dữ liệu

```python
dataFrame = dataFrame.drop(columns=["Flow ID"])
```

### 2.7. Xử lý các cột có tất cả các giá trị đều giống nhau
Loại bỏ các cột có giá trị giống nhau để tránh thừa thông tin không cần thiết

```python
dataFrame = dataFrame.loc[:, dataFrame.nunique() > 1]
dataFrame.head()
```

### 2.8. Kiểm tra dữ liệu trước khi xuất ra file
Kiểm tra dữ liệu sau khi xử lý, đồng thời xóa cột Unnamed : 0 do cột này không có giá trị phân tích

```python
dataFrame = dataFrame.drop(columns=["Unnamed: 0"])
```

```python
dataFrame.describe()
```

```python
print("Is not null:", dataFrame.notna().all().all()) # Kiểm tra null trước khi lưu dữ liệu
```

### 2.9. Lưu tập dữ liệu dưới dạng file parquet
